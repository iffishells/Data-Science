```python
df = pd.read_csv("Data/Stock_Exchange.csv")
```

```python
df.head()

## we are only interesting in open col , high col, low col,
```

```python
df = df[["Open" ,"High", "Low", "Close", "Volume"]]
df.head()
```

```python
# lets go head transfer our Data
# HL_PCT  is high low percentage
df["HL_PCT"] = (df["High"] - df["Low"]) / df["Close"] * 100
df.head()
```

```python
# next we will do daily percent Change
# PCT_change is percentage change
df["PCT_change"] = (df["Close"]-df["Open"]) / df["Open"] * 100
df.head()
```

```python
# Now we will define a new dataframe as :
df = df[["Close","HL_PCT","PCT_change","Volume"]]
df.head()
```
